#### Imports

```python
import gradio as gr
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
model_checkpoint = 'google/flan-t5-large'
```

#### Prepare dataset and Summarize

```python
model = AutoModelForSeq2SeqLM.from_pretrained(model_checkpoint)
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
```

```python
min_target_length = 32
max_target_length = 256
prefix = 'summarize: '
```

```python
def summarization(input):
  # Create string to pass to tokenizer
  summarize_input = prefix + input
  # Tokenize input
  tokenized_input = tokenizer.encode(summarize_input, return_tensors='pt')
  # Generare tokenized summary output
  tokenized_output = model.generate(tokenized_input, min_length=min_target_length, max_length=max_target_length, early_stopping=True, num_beams=5, temperature=0.8)
  # Decode tokenized summary
  summary = tokenizer.decode(tokenized_output[0], skip_special_tokens=True)

  return summary
```

```python
input_text = "The COVID-19 pandemic has had a significant impact on the global economy, leading to widespread job losses and business closures. Governments around the world have implemented various measures to support individuals and businesses during this challenging time, including financial aid, tax relief, and regulatory changes. In some cases, these measures have been successful in mitigating the worst effects of the pandemic, but they have also put a strain on government budgets and led to rising debt levels. As the world continues to grapple with the pandemic, policymakers are faced with difficult choices about how best to balance public health and economic concerns."
```

```python
summarized_output = summarization(input_text)
summarized_output
```

#### Chatgpt-like interface using gradio

```python
title = 'Chat GPT like interface using Hugging Face t5-large model checkpoint'

interface = gr.Interface(fn=summarization, inputs=gr.inputs.Textbox(lines=10, label='Input Box'), outputs='text', title=title).launch()
```
